```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Definition of SimpleNN class
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()

        self.network = nn.Sequential(
            nn.Linear(1, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )

    def forward(self, x):
        return self.network(x)

# Definitions of generate_data and plot_losses functions
def generate_data():
    torch.manual_seed(42)
    X = torch.linspace(-1, 1, 100).unsqueeze(1)
    y = 3 * X + 2 + 0.2 * torch.randn(X.size())
    return X, y

def plot_losses(train_losses, val_losses):
    plt.plot(train_losses, label="Train Loss")
    plt.plot(val_losses, label="Validation Loss")
    plt.legend()
    plt.title("Training vs Validation Loss")
    plt.show()

# Generate Data
X, y = generate_data()

# Train/Val Split
X_train, y_train = X[:80], y[:80]
X_val, y_val = X[80:], y[80:]

# Initialize Model
model = SimpleNN()

# Choose Optimizer (Switch Here)
optimizer = optim.Adam(model.parameters(), lr=0.01)
# optimizer = optim.SGD(model.parameters(), lr=0.01)
# optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

loss_fn = nn.MSELoss()

train_losses = []
val_losses = []

epochs = 200

for epoch in range(epochs):

    model.train()
    pred_train = model(X_train)
    train_loss = loss_fn(pred_train, y_train)

    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()

    model.eval()
    with torch.no_grad():
        pred_val = model(X_val)
        val_loss = loss_fn(pred_val, y_val)

    train_losses.append(train_loss.item())
    val_losses.append(val_loss.item())

    if epoch % 20 == 0:
        print(f"Epoch {epoch} | Train Loss: {train_loss.item():.4f} | Val Loss: {val_loss.item():.4f}")

plot_losses(train_losses, val_losses)
```
